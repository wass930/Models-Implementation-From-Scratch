```python
import pandas as pd
import numpy as np
```

```python
class LinearRegression:
    def __init__(self, n_iterations=100, learning_rate=0.01):
        self._n_iterations = n_iterations
        self._learning_rate = learning_rate
    
    def fit(self, X, y):
        self.m, self.n = X.shape
        self.W = np.random.rand(self.n+1)
        self.X = np.hstack((X,np.ones((self.m,1))))
        self.y = y
    
        for i in range(self._n_iterations):
            y_pred = self.predict(self.X)
            self.update_weights(y_pred)
        return self
            
    def predict(self, X):
        return self.X.dot(self.W)
    
    def score(self, X, y):
        return 1 - (np.sum((self.predict(X)-y)**2) / np.sum((y-y.mean())**2))        
    
    def update_weights(self, y_pred):
        cost = np.sum((y_pred-self.y)**2) / 2*self.m
        dW = self.X.T.dot(y_pred-y)
        self.W -= self._learning_rate * dW
        return self
```

```python
from sklearn.datasets import make_regression

X, y = make_regression(n_features=1, noise=20)
y
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
lr.predict(X)
```

```python
lr.score(X,y)
```
